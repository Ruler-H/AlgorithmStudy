# 백준 1107번 : 리모컨

```python
import sys
# input = sys.stdin.readline

# 입력
# channel : 이동하려는 채널
channel = int(input().strip())
# n : 고장난 버튼의 수
n = int(input().strip())
# button_list : 사용 가능한 버튼 리스트
button_list = [str(i) for i in range(10)]

# 고장난 버튼이 있는 경우 사용 가능 버튼 리스트에서 해당 버튼을 제외
if n > 0:
    for i in map(str, input().strip().split()):
        button_list.remove(i)
# print(button_list)
# 초기 첫번째 자리에 들어올 수를 리스트에 초기화
first_digit = int(str(channel)[0])
channel_list = []
for i in range(0, 10):
    if str(first_digit - i) in button_list:
        channel_list.append(str(first_digit - i))
    if str(first_digit + i) in button_list:
        channel_list.append(str(first_digit + i))
    if channel_list:
        break
channel_list = list(set(channel_list))
# print(channel_list)
# channel 자릿 수와 같은 자릿 수의 가능한 모든 채널을 생성
for _ in range(len(str(channel)) - 1):
    new_list = []
    for i in channel_list:
        for j in button_list:
            new_list.append(i + j)

    channel_list = new_list
# print(channel_list)
# channel 자릿 수 보다 한 자리 적은 채널 중에 가장 큰 채널과
# channel 자릿 수 보다 한 자리 큰 채널 중에 가장 큰 작은 채널을 리스트에 추가
if len(channel_list) > 1:
    channel_list.append(button_list[-1] * (len(str(channel)) - 1))
# print(channel_list)
if button_list[0] != '0':
    channel_list.append(button_list[0] * (len(str(channel)) + 1))
elif len(button_list) > 1:
    channel_list.append(button_list[1] * (len(str(channel)) + 1))

result = min(map(lambda x : (abs(channel - int(x)) + len(x)), channel_list))

if channel == 100:
    print(0)
else:
    print(min(result, abs(channel - 100)))
```
