```python
# 백준 16503번 - 괄호 없는 사칙연산
import sys
# input = sys.stdin.readline

k1, op1, k2, op2, k3 = input().split()
k1, k2, k3 = int(k1), int(k2), int(k3)
if op1 == '+':
    if op2 == '+':
        print(k1 + k2 + k3)
        print(k1 + k2 + k3)
    elif op2 == '-':
        rs1 = (k1 + k2) - k3
        rs2 = k1 + (k2 - k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '*':
        rs1 = (k1 + k2) * k3
        rs2 = k1 + (k2 * k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '/':
        rs1 = int((k1 + k2) / k3)
        rs2 = k1 + int(k2 / k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
elif op1 == '-':
    if op2 == '+':
        rs1 = (k1 - k2) + k3
        rs2 = k1 - (k2 + k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '-':
        rs1 = (k1 - k2) - k3
        rs2 = k1 - (k2 - k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '*':
        rs1 = (k1 - k2) * k3
        rs2 = k1 - (k2 * k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '/':
        rs1 = int((k1 - k2) / k3)
        rs2 = k1 - int(k2 / k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
elif op1 == '*':
    if op2 == '+':
        rs1 = (k1 * k2) + k3
        rs2 = k1 * (k2 + k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '-':
        rs1 = (k1 * k2) - k3
        rs2 = k1 * (k2 - k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '*':
        print(k1 * k2 * k3)
        print(k1 * k2 * k3)
    elif op2 == '/':
        rs1 = int((k1 * k2) / k3)
        rs2 = k1 * int(k2 / k3)
        print(min(rs1, rs2))
        print(max(rs1, rs2))
elif op1 == '/':
    if op2 == '+':
        rs1 = int(k1 / k2) + k3
        rs2 = int(k1 / (k2 + k3))
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '-':
        rs1 = int(k1 / k2) - k3
        rs2 = int(k1 / (k2 - k3))
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '*':
        rs1 = int(k1 / k2) * k3
        rs2 = int(k1 / (k2 * k3))
        print(min(rs1, rs2))
        print(max(rs1, rs2))
    elif op2 == '/':
        rs1 = int(int(k1 / k2) / k3)
        rs2 = int(k1 / int(k2 / k3))
        print(min(rs1, rs2))
        print(max(rs1, rs2))
```

```python
# 백준 25285번 - 심준의 병역판정검사
import sys
# input = sys.stdin.readline

t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    bmi = m / (n / 100) ** 2
    if n < 140.1:
        print('6')
    elif n < 146:
        print('5')
    elif n < 159:
        print('4')
    elif n < 161:
        if bmi < 16 or bmi >= 35:
            print('4')
        else:
            print('3')
    elif n < 204:
        if 20 <= bmi < 25:
            print('1')
        elif 18.5 <= bmi < 20 or 25 <= bmi < 30:
            print('2')
        elif 16 <= bmi < 18.5 or 30 <= bmi < 35:
            print('3')
        else:
            print('4')
    else:
        print('4')
```
