```python
# 백준 2915번 - 로마 숫자 재배치
import sys
# input = sys.stdin.readline

b = input().strip()
I_cnt = b.count("I")
V_cnt = b.count("V")
X_cnt = b.count("X")
C_cnt = b.count("C")
L_cnt = b.count("L")

tens_digit = ""
if C_cnt:
    if X_cnt == 1:
        tens_digit = "XC"
    elif X_cnt == 2:
        tens_digit += "XCX"
elif L_cnt:
    if X_cnt == 0:
        tens_digit = "L"
    elif X_cnt == 1:
        tens_digit = "XL"
    elif X_cnt == 2:
        tens_digit = "LXX"
    elif X_cnt == 3:
        tens_digit = "LXXX"
    elif X_cnt == 4:
        tens_digit = "LXXXX"
else:
    if X_cnt == 0:
        tens_digit = ""
    elif X_cnt == 1:
        tens_digit = "X"
    elif X_cnt == 2:
        tens_digit = "XX"
    elif X_cnt == 3:
        tens_digit = "XXX"
    elif X_cnt == 4:
        tens_digit = "XXXX"

ones_digit = ""
if V_cnt:
    if I_cnt == 0:
        ones_digit = "V"
    elif I_cnt == 1:
        ones_digit = "IV"
    elif I_cnt == 2:
        ones_digit = "VII"
    elif I_cnt == 3:
        ones_digit = "VIII"
elif I_cnt:
    if I_cnt == 1:
        ones_digit = "I"
        if tens_digit and tens_digit[-1] == "X":
            tens_digit = tens_digit[:-1]
            ones_digit += "X"
            if tens_digit == "LX":
                tens_digit = "XL"
    elif I_cnt == 2:
        ones_digit = "II"
    elif I_cnt == 3:
        ones_digit = "III"
    
print(tens_digit + ones_digit)
```
