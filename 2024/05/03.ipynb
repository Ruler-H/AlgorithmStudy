```python
# 백준 3085번 - 사탕 게임
import sys
# input = sys.stdin.readline

n = int(input())

board = []
for _ in range(n):
    board.append(list(input().strip()))
result = 0
for i in range(n):
    for j in range(n):
        is_swapped = False
        cnt = 1
        for k in range(j + 1, n):
            if board[i][j] != board[i][k]:
                if not is_swapped:
                    if i != 0 and board[i - 1][k] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif i != n - 1 and board[i + 1][k] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif k != n - 1 and board[i][k + 1] == board[i][j]:
                        cnt += 1
                        break
                    else:
                        break
                else:
                    break
            else:
                cnt += 1
        result = max(result, cnt)
        is_swapped = False
        cnt = 1
        for k in range(j - 1, -1, -1):
            if board[i][j] != board[i][k]:
                if not is_swapped:
                    if i != 0 and board[i - 1][k] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif i != n - 1 and board[i + 1][k] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif k != 0 and board[i][k - 1] == board[i][j]:
                        cnt += 1
                        break
                    else:
                        break
                else:
                    break
            else:
                cnt += 1
        result = max(result, cnt)
        is_swapped = False
        cnt = 1
        for k in range(i + 1, n):
            if board[i][j] != board[k][j]:
                if not is_swapped:
                    if j != 0 and board[k][j - 1] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif j != n - 1 and board[k][j + 1] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif k != n - 1 and board[k + 1][j] == board[i][j]:
                        cnt += 1
                        break
                    else:
                        break
                else:
                    break
            else:
                cnt += 1
        result = max(result, cnt)
        is_swapped = False
        cnt = 1
        for k in range(i - 1, -1, -1):
            if board[i][j] != board[k][j]:
                if not is_swapped:
                    if j != 0 and board[k][j - 1] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif j != n - 1 and board[k][j + 1] == board[i][j]:
                        cnt += 1
                        is_swapped = True
                    elif k != 0 and board[k - 1][j] == board[i][j]:
                        cnt += 1
                        break
                    else:
                        break
                else:
                    break
            else:
                cnt += 1
        result = max(result, cnt)
print(result)
```
