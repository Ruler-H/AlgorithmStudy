```python
# 백준 2166번 - 다각형의 면적
import sys
# input = sys.stdin.readline

N = int(input())
points = [list(map(int, input().split())) for _ in range(N)]
```

```python
result = 0

for i in range(N-1):
    result += points[i][0] * points[i+1][1] - points[i][1] * points[i+1][0]

result += points[N-1][0] * points[0][1] - points[N-1][1] * points[0][0]
result = abs(result) / 2
print("%.1f" % result)
```
