```python
# 백준 2251번 - 물통
import sys
# input = sys.stdin.readline

A, B, C = map(int, input().split())
```

```python
result = []
visited = [[0] * 201 for _ in range(201)]
visited[0][0] = 1

def bfs():
    queue = [(0, 0 , C)]
    result.append(C)

    while queue:
        a, b, c = queue.pop(0)
        # A -> B
        if a + b > B:
            if visited[a + b - B][B] == 0:
                visited[a + b - B][B] = 1
                queue.append((a + b - B, B, c))
        else:
            if visited[0][a + b] == 0:
                visited[0][a + b] = 1
                queue.append((0, a + b, c))
                result.append(c)

        # A -> C
        if a + c > C:
            if visited[a + c - C][b] == 0:
                visited[a + c - C][b] = 1
                queue.append((a + c - C, b, C))
        else:
            if visited[0][b] == 0:
                visited[0][b] = 1
                queue.append((0, b, a + c))
                result.append(a + c)

        # B -> A
        if b + a > A:
            if visited[A][b + a - A] == 0:
                visited[A][b + a - A] = 1
                queue.append((A, b + a - A, c))
        else:
            if visited[b + a][0] == 0:
                visited[b + a][0] = 1
                queue.append((b + a, 0, c))

        # B -> C
        if b + c > C:
            if visited[a][b + c - C] == 0:
                visited[a][b + c - C] = 1
                queue.append((a, b + c - C, C))
                if a == 0:
                    result.append(C)
        else:
            if visited[a][0] == 0:
                visited[a][0] = 1
                queue.append((a, 0, b + c))
                if a == 0:
                    result.append(b + c)

        # C -> A
        if c + a > A:
            if visited[A][b] == 0:
                visited[A][b] = 1
                queue.append((A, b, c + a - A))
        else:
            if visited[c + a][b] == 0:
                visited[c + a][b] = 1
                queue.append((c + a, b, 0))
        # C -> B
        if c + b > B:
            if visited[a][B] == 0:
                visited[a][B] = 1
                queue.append((a, B, c + b - B))
                if a == 0:
                    result.append(c + b - B)
        else:
            if visited[a][c + b] == 0:
                visited[a][c + b] = 1
                queue.append((a, c + b, 0))
                if a == 0:
                    result.append(0)

bfs()
result = sorted(list(set(result)))
print(" ".join(map(str, result)))
```
