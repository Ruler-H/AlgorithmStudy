```python
# 백준 11559번 - Puyo Puyo
import sys
# input = sys.stdin.readline
from collections import deque
sys.setrecursionlimit(10**6)

# ......
# ..YY..
# .YGB..
# .BGY..
# .BGY..
# .BYB..
# .GBY..
# .GBBB.
# .BGGG.
# .YGGY.
# .YGGY.
# .YBBB.
field = [
    list("......"),
    list("..YY.."),
    list(".YGB.."),
    list(".BGY.."),
    list(".BGY.."),
    list(".BYB.."),
    list(".GBY.."),
    list(".GBBB."),
    list(".BGGG."),
    list(".YGGY."),
    list(".YGGY."),
    list(".YBBB.")
]
```

```python
dxy = [(0, 1), (0, -1), (1, 0), (-1, 0)]
def bfs(x, y):
    q = deque()
    q.append((x, y))
    visited = [[False] * 6 for _ in range(12)]
    visited[x][y] = True
    cnt = 1
    while q:
        x, y = q.popleft()
        for dx, dy in dxy:
            nx, ny = x + dx, y + dy
            if 0 <= nx < 12 and 0 <= ny < 6 and not visited[nx][ny] and field[nx][ny] == field[x][y]:
                visited[nx][ny] = True
                q.append((nx, ny))
                cnt += 1
    if cnt >= 4:
        for i in range(12):
            for j in range(6):
                if visited[i][j]:
                    field[i][j] = "."
        return True
    return False

def gravity():
    for j in range(6):
        temp = []
        for i in range(12):
            if field[i][j] != ".":
                temp.append(field[i][j])
                field[i][j] = "."
        for i in range(12 - len(temp), 12):
            field[i][j] = temp[i - 12 + len(temp)]

def solution():
    cnt = 0
    while True:
        flag = False
        for i in range(12):
            for j in range(6):
                if field[i][j] != ".":
                    if bfs(i, j):
                        flag = True
        if not flag:
            break
        gravity()
        cnt += 1
    print(cnt)
    
solution()
```
