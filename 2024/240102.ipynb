```python
# 백준 25629번 - 홀짝 수열
import sys
# input = sys.stdin.readline

n = int(input())
arr = list(map(int, input().split()))
odd = 0
even = 0
for i in arr:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
if odd - even == 1 or odd - even == 0:
    print(1)
else:
    print(0)
```

```python
# 백준 17944번 - 퐁당퐁당 1
import sys
# input = sys.stdin.readline

n, t = map(int, input().split())

arr = []
for i in range(1, (n * 2) + 1):
    arr.append(i)
for i in range((n * 2) - 1, 1, -1):
    arr.append(i)

print(arr[(t % ((n * 4) - 2)) - 1])
```

```python
# 백준 15923번 - 욱제는 건축왕이야!!
import sys
# input = sys.stdin.readline

n = int(input())
arr = []
for _ in range(n):
    arr.append(list(map(int, input().split())))
result = 0
for i in range(n - 1):
    result += abs(arr[i][0] - arr[i + 1][0]) + abs(arr[i][1] - arr[i + 1][1])
result += abs(arr[0][0] - arr[n - 1][0]) + abs(arr[0][1] - arr[n - 1][1])
print(result)
```

```python
# 백준 20410번 - 추첨상 사수 대작전! (Easy)
import sys
# input = sys.stdin.readline
import itertools

m, s, x1, x2 = map(int, input().split())

arr = itertools.product(range(m), range(m))
for i in arr:
    if (i[0] * s + i[1]) % m == x1 and (i[0] * x1 + i[1]) % m == x2:
        print(i[0], i[1])
        break
```

```python
# 백준 25630번 - 팰린드롬 소떡소떡
import sys
# input = sys.stdin.readline

n = int(input())
inp_str = input().strip()

result = 0
for i in range(n // 2):
    if inp_str[i] != inp_str[n - i - 1]:
        result += 1
print(result)
```

```python
# 백준 25641번 - 균형 잡힌 소떡소떡
import sys
# input = sys.stdin.readline

n = int(input())
inp_str = input().strip()

result = ''
for i in range(n, -1, -1):
    if inp_str[i:].count('s') == inp_str[i:].count('t'):
        result = inp_str[i:]
print(result)
```

```python
# 백준 25642번 - 젓가락 게임
import sys
# input = sys.stdin.readline

a, b = map(int, input().split())

i = 0
while True:
    if i == 0:
        b += a
        i = 1
    else:
        a += b
        i = 0

    if a >= 5:
        print('yj')
        break
    elif b >= 5:
        print('yt')
        break
```
