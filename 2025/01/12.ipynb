```python
# 백준 11049번 - 행렬 곱셈 순서
import sys
# input = sys.stdin.readline
sys.setrecursionlimit(10**6)

n = int(input())
matrix = []
for _ in range(n):
    matrix.append(list(map(int, input().split())))
```

```python
dp = [[0] * n for _ in range(n)]

def solve(i, j):
    if i == j:
        return 0
    if dp[i][j]:
        return dp[i][j]
    if i+1 == j:
        return matrix[i][0] * matrix[i][1] * matrix[j][1]
    dp[i][j] = 2**32
    for k in range(i, j):
        dp[i][j] = min(dp[i][j], solve(i, k) + solve(k+1, j) + matrix[i][0] * matrix[k][1] * matrix[j][1])
    return dp[i][j]

print(solve(0, n-1))
```
