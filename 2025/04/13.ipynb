```python
import sys
from collections import deque

input = sys.stdin.readline

M, N = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(M)]

# 입력 좌표와 방향 (1-based → 0-based 조정)
cur_x, cur_y, cur_dir = map(int, input().split())
cur_x -= 1
cur_y -= 1
cur_dir -= 1
dest_x, dest_y, dest_dir = map(int, input().split())
dest_x -= 1
dest_y -= 1
dest_dir -= 1

# 방향: 동, 서, 남, 북 → 인덱스: 0, 1, 2, 3
# 문제 기준: 1=동, 2=서, 3=남, 4=북 → 인덱스: 0, 1, 2, 3
dxy = [(0, 1), (0, -1), (1, 0), (-1, 0)]

# visited[x][y][dir]
visited = [[[False]*4 for _ in range(N)] for _ in range(M)]

def turn_cost(from_dir, to_dir):
    if from_dir == to_dir:
        return 0
    elif (from_dir + to_dir == 1) or (from_dir + to_dir == 5):  # 동<->서, 남<->북
        return 2
    else:
        return 1

def bfs():
    queue = deque()
    queue.append((cur_x, cur_y, cur_dir, 0))
    visited[cur_x][cur_y][cur_dir] = True

    while queue:
        x, y, dir, cnt = queue.popleft()

        if (x, y, dir) == (dest_x, dest_y, dest_dir):
            return cnt

        # 1. 직진: 1~3칸
        dx, dy = dxy[dir]
        for k in range(1, 4):  # 최대 3칸
            nx = x + dx * k
            ny = y + dy * k
            if not (0 <= nx < M and 0 <= ny < N):
                break
            if board[nx][ny] == 1:
                break
            if not visited[nx][ny][dir]:
                visited[nx][ny][dir] = True
                queue.append((nx, ny, dir, cnt + 1))

        # 2. 방향 회전
        for new_dir in range(4):
            if new_dir != dir:
                cost = turn_cost(dir, new_dir)
                if not visited[x][y][new_dir]:
                    visited[x][y][new_dir] = True
                    queue.append((x, y, new_dir, cnt + cost))

result = bfs()
print(result)
```
