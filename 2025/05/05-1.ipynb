```python
# 백준 17779번 - 게리맨더링 2
import sys
# input = sys.stdin.readline

import sys
input = sys.stdin.readline

N = 8
# 1 2 3 4 5 6 7 8
# 2 3 4 5 6 7 8 9
# 3 4 5 6 7 8 9 1
# 4 5 6 7 8 9 1 2
# 5 6 7 8 9 1 2 3
# 6 7 8 9 1 2 3 4
# 7 8 9 1 2 3 4 5
# 8 9 1 2 3 4 5 6
A = [[1, 2, 3, 4, 5, 6, 7, 8],
      [2, 3, 4, 5, 6, 7, 8, 9],
      [3, 4, 5, 6, 7, 8, 9, 1],
      [4, 5, 6, 7, 8, 9, 1, 2],
      [5, 6, 7, 8, 9, 1, 2, 3],
      [6, 7, 8, 9, 1, 2, 3, 4],
      [7, 8, 9, 1, 2, 3, 4, 5],
      [8, 9, 1, 2, 3, 4, 5, 6]]
```

```python
def is_valid(x, y, d1, d2, N):
    # 기준점 (x, y)와 경계의 길이 d1, d2를 정한다.
    #   d1, d2 ≥ 1
    #   1 ≤ x < x+d1+d2 ≤ N
    #   1 ≤ y-d1 < y < y+d2 ≤ N
    if x + d1 + d2 <= N and y - d1 >= 1 and y + d2 <= N:
        return True
    return False

result = 1e9
total_population = sum(sum(row) for row in A)

for x in range(N):
    for y in range(N):
        for d1 in range(1, N):
            for d2 in range(1, N):
                if is_valid(x, y, d1, d2, N):
                    # 5번 선거구의 상, 하, 좌, 우 좌표 구하기
                    # 5번 선거구의 좌표는 상 = (x, y), 하 = (x + d1 + d2, y - d1 + d2), 좌 = (x+d1, y-d1), 우 = (x+d2, y+d2)이다.
                    top = (x, y)
                    bottom = (x + d1 + d2, y - d1 + d2)
                    left = (x + d1, y - d1)
                    right = (x + d2, y + d2)
                    if bottom[0] < 0 or bottom[0] >= N or bottom[1] < 0 or bottom[1] >= N or left[0] < 0 or left[0] >= N or left[1] < 0 or left[1] >= N or right[0] < 0 or right[0] >= N or right[1] < 0 or right[1] >= N:
                        continue
                    # condition = x == 2 and y == 4 and d1 == 2 and d2 == 1
                    # if condition:
                    #     print(f"기준점: ({x}, {y}), d1: {d1}, d2: {d2}")
                    #     print(f"5번 선거구 좌표: 상({top}), 하({bottom}), 좌({left}), 우({right})")
                    
                    populations = [0] * 5
                    # 1번 선거구 인구 수 구하기기
                    for i in range(top[0]):
                        for j in range(top[1] + 1):
                            # if condition:
                            #     print(f"1번 선거구 좌표: ({i}, {j})")
                            populations[0] += A[i][j]
                    cnt = 0
                    for i in range(top[0], left[0]):
                        for j in range(top[1] - cnt):
                            # if condition:
                            #     print(f"1번 선거구 좌표: ({i}, {j})")
                            populations[0] += A[i][j]
                        cnt += 1
                    # if condition:
                    #     print(f"1번 선거구 인구: {populations[0]}")

                    # 2번 선거구 인구 수 구하기기
                    for i in range(top[0]):
                        for j in range(top[1] + 1, N):
                            # if condition:
                            #     print(f"2번 선거구 좌표: ({i}, {j})")
                            populations[1] += A[i][j]
                    cnt = 0
                    for i in range(top[0], right[0] + 1):
                        for j in range(top[1] + cnt + 1, N):
                            # if condition:
                            #     print(f"2번 선거구 좌표: ({i}, {j})")
                            populations[1] += A[i][j]
                        cnt += 1
                    # if condition:
                    #     print(f"2번 선거구 인구: {populations[1]}")

                    # 3번 선거구 인구 수 구하기기
                    cnt = 0
                    for i in range(left[0], bottom[0] + 1):
                        for j in range(left[1] + cnt):
                            # if condition:
                            #     print(f"3번 선거구 좌표: ({i}, {j})")
                            try:
                                populations[2] += A[i][j]
                            except IndexError:
                                print(f"IndexError: ({i}, {j})")
                        cnt += 1

                    for i in range(bottom[0] + 1, N):
                        for j in range(bottom[1]):
                            # if condition:
                            #     print(f"3번 선거구 좌표: ({i}, {j})")
                            populations[2] += A[i][j]

                    # if condition:
                    #     print(f"3번 선거구 인구: {populations[2]}")

                    # 4번 선거구 인구 수 구하기기
                    cnt = 0
                    for i in range(right[0] + 1, bottom[0] + 1):
                        for j in range(right[1] - cnt, N):
                            # if condition:
                            #     print(f"4번 선거구 좌표: ({i}, {j})")
                            populations[3] += A[i][j]
                        cnt += 1

                    for i in range(bottom[0] + 1, N):
                        for j in range(bottom[1], N):
                            # if condition:
                            #     print(f"4번 선거구 좌표: ({i}, {j})")
                            populations[3] += A[i][j]
                    # if condition:
                    #     print(f"4번 선거구 인구: {populations[3]}")
                    
                    # 5번 선거구 인구 수 구하기기
                    populations[4] = total_population - sum(populations)

                    # if condition:
                    #     print(f"인구 수: {populations}")

                    # 최대, 최소 인구 수 구한 후 차이 구하기
                    max_population = max(populations)
                    min_population = min(populations)
                    diff = max_population - min_population
                    result = min(result, diff)

print(result)
```
