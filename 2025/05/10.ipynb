```python
# 백준 23288번 - 주사위 굴리기 2
import sys
from collections import deque
# input = sys.stdin.readline

N, M, K = map(int, input().split())
board = [list(map(int, input().split())) for _ in range(N)]
```

```python
# 동, 남, 서, 북 (시계방향)
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]
direction = 0  # 처음 방향: 동쪽

# 주사위 상태
dice = {
    'top': 1,
    'bottom': 6,
    'north': 2,
    'south': 5,
    'east': 3,
    'west': 4
}

x, y = 0, 0
score = 0

def move_dice(d):
    t, b, n, s, e, w = dice['top'], dice['bottom'], dice['north'], dice['south'], dice['east'], dice['west']
    if d == 0:  # 동쪽
        dice['top'] = w
        dice['bottom'] = e
        dice['east'] = t
        dice['west'] = b
    elif d == 1:  # 남쪽
        dice['top'] = n
        dice['bottom'] = s
        dice['north'] = b
        dice['south'] = t
    elif d == 2:  # 서쪽
        dice['top'] = e
        dice['bottom'] = w
        dice['east'] = b
        dice['west'] = t
    elif d == 3:  # 북쪽
        dice['top'] = s
        dice['bottom'] = n
        dice['north'] = t
        dice['south'] = b

def get_score(sx, sy):
    visited = [[False]*M for _ in range(N)]
    visited[sx][sy] = True
    q = deque()
    q.append((sx, sy))
    cnt = 1
    val = board[sx][sy]
    
    while q:
        x, y = q.popleft()
        for d in range(4):
            nx, ny = x + dx[d], y + dy[d]
            if 0 <= nx < N and 0 <= ny < M and not visited[nx][ny] and board[nx][ny] == val:
                visited[nx][ny] = True
                cnt += 1
                q.append((nx, ny))
    return cnt * val

def rotate_direction(curr_dir, bottom, cell_val):
    if bottom > cell_val:
        return (curr_dir + 1) % 4  # 시계
    elif bottom < cell_val:
        return (curr_dir - 1) % 4  # 반시계
    else:
        return curr_dir  # 그대로

for _ in range(K):
    # 다음 좌표 계산
    nx, ny = x + dx[direction], y + dy[direction]

    # 경계 밖이면 방향 반대로 바꾸고 이동
    if not (0 <= nx < N and 0 <= ny < M):
        direction = (direction + 2) % 4
        nx, ny = x + dx[direction], y + dy[direction]

    # 주사위 이동 및 점수 획득
    move_dice(direction)
    x, y = nx, ny
    score += get_score(x, y)

    # 방향 회전
    direction = rotate_direction(direction, dice['bottom'], board[x][y])

print(score)
```
