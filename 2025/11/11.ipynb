```python
# 백준 32710번 - 구구단표
import sys
# input = sys.stdin.readline

N = int(input())
flag = False
for i in range(1, 10):
    for j in range(1, 10):
        if N == i or N == j or N == i * j:
            print("1")
            flag = True
            break
    if flag:
        break
if not flag:
    print("0")
```
