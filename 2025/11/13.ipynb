```python
# 백준 15751번 - Teleportation
import sys
# input = sys.stdin.readline

a, b, x, y = map(int, input().split())

# 옵션 1: 텔레포터 미사용
direct = abs(b - a)

# 옵션 2: a→x로 이동, 텔레포터로 y로 이동, y→b로 이동
with_teleporter1 = abs(x - a) + abs(b - y)

# 옵션 3: a→y로 이동, 텔레포터로 x로 이동, x→b로 이동
with_teleporter2 = abs(y - a) + abs(b - x)

print(min(direct, with_teleporter1, with_teleporter2))
```
