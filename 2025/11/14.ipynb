```python
# 백준 28323번 - 불안정한 수열
import sys
# input = sys.stdin.readline

n = int(input())
a = list(map(int, input().split()))

max_even = 0  # 짝수로 끝나는 최대 길이
max_odd = 0   # 홀수로 끝나는 최대 길이

for num in a:
    if num % 2 == 0:  # 짝수
        new_even = max(1, max_odd + 1)
        max_even = max(max_even, new_even)
    else:  # 홀수
        new_odd = max(1, max_even + 1)
        max_odd = max(max_odd, new_odd)

answer = max(max_even, max_odd)
print(answer)
```
