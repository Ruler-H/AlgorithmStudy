```python
# 백준 18512번 - 점프 점프
import sys
# input = sys.stdin.readline
import math

X, Y, P1, P2 = map(int, input().split())

g = math.gcd(X, Y)

# 공통 지점이 존재하는지 확인
if (P2 - P1) % g != 0:
    print(-1)
else:
    # P1 + n*X >= P2를 만족하는 최소 n
    n_min = max(0, -(-max(0, P2 - P1) // X))  # ceil 대신 사용
    
    found = False
    for n in range(n_min, n_min + Y // g):
        pos = P1 + n * X
        if (pos - P2) % Y == 0:
            print(pos)
            found = True
            break
    
    if not found:
        print(-1)
```
