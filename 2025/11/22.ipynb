```python
# 백준 23841번 - 데칼코마니
import sys
# input = sys.stdin.readline

N, M = map(int, input().split())

picture = [["."] * M for _ in range(N)]

for i in range(N):
    inp_pic = input().strip()
    for j in range(M):
        if inp_pic[j] != ".":
            picture[i][j] = inp_pic[j]
            picture[i][M - j - 1] = inp_pic[j]

for i in range(N):
    print("".join(picture[i]))
```
