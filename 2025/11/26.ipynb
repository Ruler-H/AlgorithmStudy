```python
# 백준 30794번 - 가희와 클럽 오디션 1
import sys
# input = sys.stdin.readline

l, verdict = map(str, input().split())
l = int(l)

if verdict == "miss":
    print(0)
elif verdict == "bad":
    print(l * 200)
elif verdict == "cool":
    print(l * 400)
elif verdict == "great":
    print(l * 600)
elif verdict == "perfect":
    print(l * 1000)
```
