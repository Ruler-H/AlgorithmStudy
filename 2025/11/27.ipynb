```python
# 백준 17247번 - 택시 거리
import sys
# input = sys.stdin.readline

N, M = map(int, input().split())

point_1 = None
point_2 = None

for i in range(N):
    row = list(map(int, input().split()))
    for j in range(M):
        if row[j] == 1:
            if point_1 is None:
                point_1 = (i, j)
            elif point_2 is None:
                point_2 = (i, j)

print(abs(point_1[0] - point_2[0]) + abs(point_1[1] - point_2[1]))
```
