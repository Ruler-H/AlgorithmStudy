```python
# 백준 23882번 - 알고리즘 수업 - 선택 정렬 2
import sys
# input = sys.stdin.readline

N, K = map(int, input().split())
A = list(map(int, input().split()))
cnt = 0

for i in range(N - 1, 0, -1):
    max_index = A.index(max(A[:i+1]))
    if max_index != i:
        A[max_index], A[i] = A[i], A[max_index]
        cnt += 1
    if cnt == K:
        print(" ".join(map(str, A)))
        break
else:
    print(-1)
```
