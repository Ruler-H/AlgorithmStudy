```python
# 백준 24389번 - 2의 보수
import sys
# input = sys.stdin.readline

N = int(input())

# N의 2진수
bin_N = bin(N)[2:].zfill(32)

# 비트 반전
not_bin_N = ''.join('1' if bit == '0' else '0' for bit in bin_N)

# 1을 더함
complement = int(not_bin_N, 2) + 1
bin_complement = bin(complement)[2:].zfill(32)

cnt = 0
for i in range(32):
    if bin_N[i] != bin_complement[i]:
        cnt += 1

print(cnt)
```
