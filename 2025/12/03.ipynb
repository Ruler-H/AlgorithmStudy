```python
# 백준 9494번 - Text Roll
import sys
# input = sys.stdin.readline

while True:
    n = int(input())
    if n == 0:
        break

    texts = [input().rstrip('\n') for _ in range(n)]

    # --- 첫 줄 처리: 첫 단어의 끝 뒤 위치 구하기 ---
    first = texts[0]

    # 첫 단어 시작 인덱스 (첫 non-space)
    start = 0
    while start < len(first) and first[start] == ' ':
        start += 1

    # 첫 단어가 끝난 뒤 공백(또는 문자열 끝)
    pos = first.find(' ', start)
    if pos == -1:
        pos = len(first)  # 줄 끝 뒤의 가상 공백

    # --- 나머지 줄 처리 ---
    for i in range(1, n):
        line = texts[i]
        if pos < len(line):
            idx = line.find(' ', pos)
            if idx == -1:
                idx = len(line)
            pos = idx
        # pos >= len(line) 이면 이미 빈 공간 위 → 그대로 아래로 내려감

    print(pos + 1)
```
