```python
# 백준 17284번 - Vending Machine
import sys
# input = sys.stdin.readline

orders = list(map(int, input().split()))

result = 5000

for order in orders:
    if order == 1:
        result -= 500
    elif order == 2:
        result -= 800
    elif order == 3:
        result -= 1000
        
print(result)
```
