```python
# 백준 15904번 - UCPC는 무엇의 약자일까?
import sys
# input = sys.stdin.readline

inp_str = input().strip()
UCPC = ["U", "C", "P", "C"]

for i in range(len(inp_str)):
    if inp_str[i] == UCPC[0]:
        UCPC.pop(0)
        if len(UCPC) == 0:
            print("I love UCPC")
            break

if len(UCPC) > 0:
    print("I hate UCPC")
```
