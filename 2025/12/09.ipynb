```python
# 백준 32978번 - 아 맞다 마늘
import sys
# input = sys.stdin.readline

N = int(input())
need_materials = list(map(str, input().split()))
need_materials.sort()
include_materials = list(map(str, input().split()))
include_materials.sort()

for i in range(N - 1):
    if need_materials[i] != include_materials[i]:
        print(need_materials[i])
        break
    if i == N - 2:
        print(need_materials[i + 1])
```
