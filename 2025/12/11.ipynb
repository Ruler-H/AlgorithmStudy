```python
# 백준 24051번 - 알고리즘 수업 - 삽입 정렬 1
import sys
# input = sys.stdin.readline

N, K = map(int, input().split())
A = [0] + list(map(int, input().split()))  # 1-indexed 배열

save_count = 0
kth_value = -1

for i in range(2, N + 1):
    loc = i - 1
    newItem = A[i]
    
    # while 루프에서 발생하는 저장들
    while 1 <= loc and newItem < A[loc]:
        A[loc + 1] = A[loc]  # 저장 발생
        save_count += 1
        if save_count == K:
            kth_value = A[loc]
        loc -= 1
    
    # 마지막 저장
    if loc + 1 != i:
        A[loc + 1] = newItem  # 저장 발생
        save_count += 1
        if save_count == K:
            kth_value = newItem

print(kth_value)
```
