```python
# 백준 13496번 - The Merchant of Venice
import sys
# input = sys.stdin.readline

k = int(input())

for i in range(k):
    n, s, d = map(int, input().split())  # d = 만기까지의 날수
    result = 0
    for _ in range(n):
        distance, value = map(int, input().split())  # 변수명 변경
        if distance <= s * d:  # 배가 d일 내에 도착하는가?
            result += value
    print(f"Data Set {i + 1}:")
    print(result)
    print()
```
