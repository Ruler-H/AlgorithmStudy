```python
# 백준 11800번 - Tawla
import sys
# input = sys.stdin.readline

T = int(input())

byelching = {
    1 : "Yakk",
    2 : "Doh",
    3 : "Seh",
    4 : "Ghar",
    5 : "Bang",
    6 : "Sheesh"
}
for i in range(T):
    nums = list(map(int, input().split()))
    nums.sort(reverse=True)
    if nums == [6, 5]:
        print(f"Case {i+1}: Sheesh Beesh")
    elif nums == [1, 1]:
        print(f"Case {i+1}: Habb Yakk")
    elif nums == [2, 2]:
        print(f"Case {i+1}: Dobara")
    elif nums == [3, 3]:
        print(f"Case {i+1}: Dousa")
    elif nums == [4, 4]:
        print(f"Case {i+1}: Dorgy")
    elif nums == [5, 5]:
        print(f"Case {i+1}: Dabash")
    elif nums == [6, 6]:
        print(f"Case {i+1}: Dosh")
    else:
        print(f"Case {i+1}: {byelching[nums[0]]} {byelching[nums[1]]}")
```
