```python
# 백준 11070번 - 피타고라스 기댓값
import sys
# input = sys.stdin.readline

T = int(input())

for _ in range(T):
    n, m = map(int, input().split())
    teams = [[0, 0] for _ in range(n)]
    for _ in range(m):
        a, b, p, q = map(int, input().split())
        teams[a-1][0] += p
        teams[a-1][1] += q
        teams[b-1][0] += q
        teams[b-1][1] += p

    result = []
    for i in range(n):
        if teams[i][0] == 0 and teams[i][1] == 0:
            result.append(0)
        else:
            result.append(int(((teams[i][0] ** 2) / ((teams[i][0] ** 2) + (teams[i][1] ** 2))) * 1000))

    print(max(result))
    print(min(result))
```
