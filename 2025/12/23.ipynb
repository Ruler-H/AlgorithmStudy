```python
# 백준 17206번 - 준석이의 수학 숙제
import sys
# input = sys.stdin.readline

def sum_multiples(n, d):
    """d의 배수 중 n 이하인 것들의 합"""
    k = n // d
    return d * k * (k + 1) // 2

T = int(input())
problems = list(map(int, input().split()))

for n in problems:
    result = sum_multiples(n, 3) + sum_multiples(n, 7) - sum_multiples(n, 21)
    print(result)
```
