```python
# 백준 10708번 - 크리스마스 파티
import sys
# input = sys.stdin.readline

N = int(input())
M = int(input())
targets = list(map(int, input().split()))
scores = [0 for _ in range(N)]
for target in targets:
    names = list(map(int, input().split()))
    for i in range(N):
        if names[i] == target:
            scores[i] += 1
        else:
            scores[target-1] += 1

for score in scores:
    print(score)
```
