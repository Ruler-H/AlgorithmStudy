```python
# 백준 28249번 - Chili Peppers
import sys
# input = sys.stdin.readline

T = int(input())

pappers = {
    "Poblano": 1500,
    "Mirasol": 6000,
    "Serrano": 15500,
    "Cayenne": 40000,
    "Thai": 75000,
    "Habanero": 125000
}
result = 0

for _ in range(T):
    pepper = input().strip()
    result += pappers[pepper]

print(result)
```
