```python
# 백준 15593번 - Lifeguards (Bronze)
import sys
# input = sys.stdin.readline

N = int(input())

time = [0 for _ in range(1000001)]
work_time = []

for i in range(N):
    s, e = map(int, input().split())
    work_time.append((s, e))
    for j in range(s, e):
        time[j] += 1

result = 0

for i in range(N):
    s, e = work_time[i]
    for j in range(s, e):
        time[j] -= 1
    imsi = 0
    for j in range(1000001):
        if time[j] > 0:
            imsi += 1
    result = max(result, imsi)
    for j in range(s, e):
        time[j] += 1

print(result)
```
