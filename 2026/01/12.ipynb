```python
# 백준 17072번 - 아스키 아트
import sys
# input = sys.stdin.readline

# 0 이상 ~ 510,000 미만	#	35
# 510,000 이상 ~ 1,020,000 미만	o (알파벳 소문자)	111
# 1,020,000 이상 ~ 1,530,000 미만	+	43
# 1,530,000 이상 ~ 2,040,000 미만	-	45
# 2,040,000 이상	.	46
intensity = ["#", "o", "+", "-", "."]
N, M = map(int, input().split())

for _ in range(N):
    rgb_list = list(map(int, input().split()))
    for i in range(0, 3 * M, 3):
        r, g, b = rgb_list[i], rgb_list[i + 1], rgb_list[i + 2]
        intensity_value = r * 2126 + g * 7152 + b * 722
        if intensity_value < 510000:
            print(intensity[0], end="")
        elif intensity_value < 1020000:
            print(intensity[1], end="")
        elif intensity_value < 1530000:
            print(intensity[2], end="")
        elif intensity_value < 2040000:
            print(intensity[3], end="")
        else:
            print(intensity[4], end="")
    print()
```
