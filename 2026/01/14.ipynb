```python
# 백준 20112번 - 사토르 마방진
import sys
# input = sys.stdin.readline

N = int(input())
words = [input().strip() for _ in range(N)]

for i in range(N):
    width = words[i]
    height = ""
    for j in range(N):
        height += words[j][i]
    if width != height:
        print("NO")
        break
else:
    print("YES")
```
