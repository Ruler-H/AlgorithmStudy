```python
# 백준 4635번 - Speed Limit
import sys
# input = sys.stdin.readline

while True:
    n = int(input())
    if n == -1:
        break
    s, t = map(int, input().split())
    prev_t = t
    result = s * t
    for _ in range(n-1):
        s, t = map(int, input().split())
        result += s * (t - prev_t)
        prev_t = t
    print(f"{result} miles")
```
