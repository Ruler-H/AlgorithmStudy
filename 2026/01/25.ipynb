```python
# 백준 9782번 - Median
import sys

i = 1
for line in sys.stdin:
    data = list(map(int, line.split()))
    n = data[0]
    
    if n == 0:
        break
    
    values = data[1:n+1]  # 정렬하지 않음
    
    if n % 2 == 1:
        median = values[n // 2]
    else:
        median = (values[n // 2 - 1] + values[n // 2]) / 2
    
    print(f"Case {i}: {float(median)}")
    i += 1
```
