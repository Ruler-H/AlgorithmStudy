```python
# 백준 31562번 - 전주 듣고 노래 맞히기
import sys
# input = sys.stdin.readline

N, M = map(int, input().split())

mp = {}  # key(3음) -> 제목 또는 '?'

for _ in range(N):
    parts = input().split()
    title = parts[1]
    key = ''.join(parts[2:5])  # a1 a2 a3
    if key not in mp:
        mp[key] = title
    else:
        mp[key] = "?"   # 이미 하나 있었다면 이제부터는 모호

for _ in range(M):
    qkey = ''.join(input().split())  # b1 b2 b3
    if qkey not in mp:
        print("!")
    else:
        print(mp[qkey])
```
