```python
# 백준 27930번 - 당신은 운명을 믿나요?
import sys
# input = sys.stdin.readline

S = input().strip()

k = 0  # KOREA 진행도
y = 0  # YONSEI 진행도

KOREA = "KOREA"
YONSEI = "YONSEI"

for ch in S:
    if k < 5 and ch == KOREA[k]:
        k += 1
        if k == 5:
            print("KOREA")
            break

    if y < 6 and ch == YONSEI[y]:
        y += 1
        if y == 6:
            print("YONSEI")
            break
```
