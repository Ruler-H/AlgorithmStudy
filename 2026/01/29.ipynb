```python
# 백준 10384번 - 팬그램
import sys
# input = sys.stdin.readline

n = int(input())

for i in range(n):
    S = input().strip()
    S = S.lower()
    cnt = [0] * 26
    for ch in S:
        if ch.isalpha():
            cnt[ord(ch) - ord("a")] += 1

    if all(c >= 3 for c in cnt):
        print(f"Case {i+1}: Triple pangram!!!")
    elif all(c >= 2 for c in cnt):
        print(f"Case {i+1}: Double pangram!!")
    elif all(c >= 1 for c in cnt):
        print(f"Case {i+1}: Pangram!")
    else:
        print(f"Case {i+1}: Not a pangram")
```
