```python
# 백준 3226번 - 전화 요금
import sys
# input = sys.stdin.readline

N = int(input())
total = 0

DAY_START = 7 * 60     # 07:00
DAY_END   = 19 * 60    # 19:00
MINUTES_IN_DAY = 24 * 60

for _ in range(N):
    t, d = input().split()
    d = int(d)
    h, m = map(int, t.split(':'))
    cur = h * 60 + m

    for _ in range(d):
        if DAY_START <= cur < DAY_END:
            total += 10
        else:
            total += 5
        cur = (cur + 1) % MINUTES_IN_DAY

print(total)
```
