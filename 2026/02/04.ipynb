```python
# 백준 33571번 - 구멍
import sys
# input = sys.stdin.readline

S = input()

one_circle = ["A", "a", "b", "D", "d", "e", "g", "o", "O", "P", "p", "q", "Q", "R", "@"]
two_circle = ["B"]

result = 0

for i in range(len(S)):
    if S[i] in one_circle:
        result += 1
    elif S[i] in two_circle:
        result += 2

print(result)
```
