```python
# 백준 5046번 - 전국 대학생 프로그래밍 대회 동아리 연합
import sys

input = sys.stdin.readline

N, B, H, W = map(int, input().split())
best = B + 1  # 예산보다 큰 값으로 초기화

for _ in range(H):
    p = int(input().strip())          # 1인당 숙박비
    weeks = list(map(int, input().split()))  # 각 주 수용 인원

    # 어떤 주라도 N명을 수용 가능하면, 해당 호텔 비용 후보
    if any(a >= N for a in weeks):
        cost = p * N
        if cost <= B and cost < best:
            best = cost

if best <= B:
    print(best)
else:
    print("stay home")
```
