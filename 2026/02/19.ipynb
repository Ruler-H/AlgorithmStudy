```python
# 백준 31306번 - Is Y a Vowel?
import sys
# input = sys.stdin.readline

inp_str = input()

vowels = ['a', 'e', 'i', 'o', 'u']
cnt = 0
include_y_cnt = 0

for i in range(len(inp_str)):
    if inp_str[i] in vowels:
        cnt += 1
        include_y_cnt += 1
    elif inp_str[i] == 'y':
        include_y_cnt += 1

print(cnt, include_y_cnt)
```
