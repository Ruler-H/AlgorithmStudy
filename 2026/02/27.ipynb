```python
# 백준 21665번 - Миша и негатив
import sys
# input = sys.stdin.readline

n, m = map(int, input().split())

pictures = [list(input().strip()) for _ in range(n)]
input()
cnt = 0
for i in range(n):
    inp_str = input().strip()
    for j in range(m):
        if pictures[i][j] == inp_str[j]:
            cnt += 1

print(cnt)
```
