```python
# 백준 25893번 - Majestic 10
import sys
# input = sys.stdin.readline

n = int(input())

for i in range(n):
    nums = map(int, input().split())
    cnt = 0
    for num in nums:
        print(num, end=" ")
        if num >= 10:
            cnt += 1
    print()
    if cnt == 0:
        print("zilch")
    elif cnt == 1:
        print("double")
    elif cnt == 2:
        print("double-double")
    elif cnt == 3:
        print("triple-double")
    print()
```
