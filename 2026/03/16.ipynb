```python
# 백준 14593번 - 2017 아주대학교 프로그래밍 경시대회 (Large)
import sys
# input = sys.stdin.readline

N = int(input())

first = 1
first_s, first_c, first_l = map(int, input().split())

for i in range(2, N + 1):
    s, c, l = map(int, input().split())
    if first_s < s:
        first = i
        first_s, first_c, first_l = s, c, l
    elif first_s == s and first_c > c:
        first = i
        first_s, first_c, first_l = s, c, l
    elif first_s == s and first_c == c and first_l > l:
        first = i
        first_s, first_c, first_l = s, c, l

print(first)
```
