```python
# 백준 14761번 - FizzBuzz
import sys
# input = sys.stdin.readline

X, Y, N = map(int, input().split())

for i in range(1, N + 1):
    if i % X == 0:
        if i % Y == 0:
            print("FizzBuzz")
        else:
            print("Fizz")
    elif i % Y == 0:
        print("Buzz")
    else:
        print(i)
```
