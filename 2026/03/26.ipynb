```python
# 백준 26742번 - Skarpetki
import sys
# input = sys.stdin.readline

b_cnt = 0
c_cnt = 0
result = 0

# 양말 입력
socks = input().strip()

# 양말 카운트
for sock in socks:
    if sock == 'B':
        b_cnt += 1
    elif sock == 'C':
        c_cnt += 1
result += (b_cnt // 2) + (c_cnt // 2)

print(result)
```
