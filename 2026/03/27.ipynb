```python
# 백준 6030번 - Scavenger Hunt
import sys
# input = sys.stdin.readline

P, Q = map(int, input().split())

def measure_operation(num):
    measure = []
    for i in range(1, int(num ** (1 / 2)) + 1):
        if num % i == 0:
            measure.append(i)
            if i != num // i:
                measure.append(num // i)
    measure.sort()
    return measure

P_measure = measure_operation(P)
Q_measure = measure_operation(Q)
for i in P_measure:
    for j in Q_measure:
        print(i, j)
```
