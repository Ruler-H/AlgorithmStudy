```python
# 백준 5698번 - Tautogram
import sys
# input = sys.stdin.readline

while True:
    inp_str = input().strip()
    if inp_str == "*":
        break
    words = inp_str.split()
    first = words[0][0].lower()
    if all(w[0].lower() == first for w in words):
        print("Y")
    else:
        print("N")
```
