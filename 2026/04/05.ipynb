```python
# 백준 34071번 - 첫 번째 문제는 정말 쉬운 문제일까?
import sys
# input = sys.stdin.readline

N = int(input())
grades = [int(input()) for _ in range(N)]

hard_grade_index = grades.index(max(grades))
easy_grade_index = grades.index(min(grades))

if hard_grade_index == 0:
    print("hard")
elif easy_grade_index == 0:
    print("ez")
else:
    print("?")
```
