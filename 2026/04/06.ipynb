```python
# 백준 17027번 - Shell Game
import sys
# input = sys.stdin.readline

N = int(input())
steps = [list(map(int, input().split())) for _ in range(N)]

best = 0
for start in [1, 2, 3]:          # 초기 조약돌 위치 3가지 시도
    pos = start
    score = 0
    for a, b, g in steps:
        if pos == a:              # swap
            pos = b
        elif pos == b:
            pos = a
        if pos == g:              # 맞췄는지 확인
            score += 1
    best = max(best, score)

print(best)
```
