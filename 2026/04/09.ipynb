```python
# 백준 14709번 - 여우 사인
import sys
# input = sys.stdin.readline

N = int(input())
pairs = set()
for _ in range(N):
    a, b = map(int, input().split())
    pairs.add((min(a, b), max(a, b)))

# 여우 사인 조건:
# 1, 3, 4가 서로 맞닿음 → (1,3), (1,4), (3,4)
# 2, 5는 어떤 쌍에도 포함되지 않음
fox = {(1, 3), (1, 4), (3, 4)}

if pairs == fox:
    print("Wa-pa-pa-pa-pa-pa-pow!")
else:
    print("Woof-meow-tweet-squeek")
```
