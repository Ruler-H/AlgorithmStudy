```python
# 백준 18868번 - 멀티버스 Ⅰ
import sys
# input = sys.stdin.readline

M, N = map(int, input().split())
universes = []
for _ in range(M):
    planets = list(map(int, input().split()))
    # 각 행성의 상대적 순위를 구한다
    sorted_unique = sorted(set(planets))
    rank_map = {v: i for i, v in enumerate(sorted_unique)}
    rank = tuple(rank_map[p] for p in planets)
    universes.append(rank)

count = 0
for i in range(M):
    for j in range(i + 1, M):
        if universes[i] == universes[j]:
            count += 1

print(count)
```
