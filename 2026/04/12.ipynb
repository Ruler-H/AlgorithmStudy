```python
# 백준 15463번 - Blocked Billboard
import sys
# input = sys.stdin.readline

# 광고판 1의 좌표
x1, y1, x2, y2 = map(int, input().split())
# 광고판 2의 좌표
x3, y3, x4, y4 = map(int, input().split())
# 트럭의 좌표
truck_x1, truck_y1, truck_x2, truck_y2 = map(int, input().split())

# 광고판 1의 넓이
result = (x2 - x1) * (y2 - y1)
# 광고판 2의 넓이
result += (x4 - x3) * (y4 - y3)

# 트럭과 첫 번째 광고판 겹치는 부분 넓이 계산
overlap_w = max(0, min(truck_x2, x2) - max(truck_x1, x1))
overlap_h = max(0, min(truck_y2, y2) - max(truck_y1, y1))
truck_area = overlap_w * overlap_h
# 트럭과 두 번째 광고판 겹치는 부분 넓이 계산
overlap_w2 = max(0, min(truck_x2, x4) - max(truck_x1, x3))
overlap_h2 = max(0, min(truck_y2, y4) - max(truck_y1, y3))
truck_area += overlap_w2 * overlap_h2

result -= truck_area

print(result)
```
