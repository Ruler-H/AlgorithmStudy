```python
# 백준 23027번 - 1번부터 문제의 상태가…?
import sys
# input = sys.stdin.readline

S = input().strip()

if "A" in S:
    for i in ["B", "C", "D", "F"]:
        S = S.replace(i, "A")
elif "B" in S:
    for i in ["C", "D", "F"]:
        S = S.replace(i, "B")
elif "C" in S:
    for i in ["D", "F"]:
        S = S.replace(i, "C")
else:
    S = "A" * len(S)

print(S)
```
